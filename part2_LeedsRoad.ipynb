```python
# Task D: Knowledge Graph Embedding + Visualization
!pip install pykeen[all]

from pykeen.pipeline import pipeline
from pykeen.datasets import CoDExMedium
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

dataset = CoDExMedium(create_inverse_triples=True)

#TransE
trans_result = pipeline(
    model='TransE',
    dataset=dataset,
    training_loop='sLCWA',
    epochs=25,
    random_seed=42
)

#RotatE
rotate_result = pipeline(
    model='RotatE',
    dataset=dataset,
    training_loop='sLCWA',
    epochs=25,
    random_seed=42
)

# ✅ 4. 训练 GCN 模型（CompGCN）
gcn_result = pipeline(
    model='CompGCN',
    dataset=dataset,
    training_loop='sLCWA',
    epochs=25,
    random_seed=42
)


def plot_embeddings(model, title):
    emb = model.entity_representations[0]().detach().cpu().numpy()

    if np.iscomplexobj(emb):
        emb = np.concatenate([emb.real, emb.imag], axis=1)

    reduced = TSNE(n_components=2, perplexity=30, random_state=42).fit_transform(emb)

    plt.figure(figsize=(10, 8))
    plt.scatter(reduced[:, 0], reduced[:, 1], s=10, alpha=0.6)
    plt.title(title)
    plt.xlabel("Dimension 1")
    plt.ylabel("Dimension 2")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_embeddings(trans_result.model, "t-SNE Visualization of TransE Entity Embeddings (CoDEx-Medium)")
plot_embeddings(rotate_result.model, "t-SNE Visualization of RotatE Entity Embeddings (CoDEx-Medium)")
plot_embeddings(gcn_result.model, "t-SNE Visualization of GCN (CompGCN) Entity Embeddings (CoDEx-Medium)")
```
