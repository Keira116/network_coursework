```python
import pandas as pd
import networkx as nx
from itertools import combinations
import matplotlib.pyplot as plt
import numpy as np

```

```python
# 1. Load Data from USERS.csv
df = pd.read_csv('USERS.csv')
print("Data Preview:")
print(df.head())

# 2. Initialize the Graph
G = nx.Graph()

# Add each unique editor as a node (assume column name is 'username')
unique_users = df['username'].unique()
for user in unique_users:
    G.add_node(user)

# 3. Construct Edges based on common participation in the same page and thread
# (Assume the CSV has 'page' and 'thread' columns)
grouped = df.groupby(['page_name', 'thread_subject'])
for (page, thread), group in grouped:
    # Get all unique editors in this group
    users_in_group = group['username'].unique()
    # For every pair of editors, create or update an edge
    for user1, user2 in combinations(users_in_group, 2):
        if G.has_edge(user1, user2):
            G[user1][user2]['weight'] += 1
            G[user1][user2]['contexts'].append((page, thread))
        else:
            G.add_edge(user1, user2, weight=1, contexts=[(page, thread)])

# Output network summary information
print("\nNetwork Summary:")
print("Number of Nodes (Editors):", G.number_of_nodes())
print("Number of Edges (Social Connections):", G.number_of_edges())

# Optional: Visualize the network (suitable for smaller networks)
plt.figure(figsize=(8,8))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=300, font_size=8)
plt.title("Wikidata Editor Social Network (from USERS.csv)")
plt.show()
```

```python
#TASKB
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
degrees = [d for _, d in G.degree()]
avg_degree = np.mean(degrees)
max_degree = np.max(degrees)
avg_clustering = nx.average_clustering(G)

print("\nReal Network Metrics:")
print("Number of nodes (editors):", num_nodes)
print("Number of edges (social connections):", num_edges)
print("Average degree:", avg_degree)
print("Maximum degree:", max_degree)
print("Average clustering coefficient:", avg_clustering)

# Analyze connectivity: find the largest connected component (LCC)
components = list(nx.connected_components(G))
print("Number of connected components:", len(components))
largest_cc = max(components, key=len)
G_lcc = G.subgraph(largest_cc).copy()
print("Size of the largest connected component:", G_lcc.number_of_nodes())

if nx.is_connected(G_lcc):
    diameter = nx.diameter(G_lcc)
    avg_shortest_path = nx.average_shortest_path_length(G_lcc)
    print("Diameter of LCC:", diameter)
    print("Average shortest path length (LCC):", avg_shortest_path)
else:
    print("Largest connected component is not fully connected.")

# Plot degree distribution (log-log scale)
plt.figure(figsize=(6,4))
hist = nx.degree_histogram(G)
plt.plot(range(len(hist)), hist, 'bo')
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Degree (log scale)")
plt.ylabel("Frequency (log scale)")
plt.title("Degree Distribution (Log-Log)")
plt.show()

# -----------------------
# 4. Compare with a Random Network
# -----------------------
n = num_nodes
# Calculate probability for ER graph so that average degree ~ avg_degree
p = avg_degree / (n - 1) if n > 1 else 0

G_random = nx.erdos_renyi_graph(n, p, seed=42)

# Compute metrics for the random network
if G_random.number_of_nodes() > 0:
    components_rand = list(nx.connected_components(G_random))
    largest_cc_rand = max(components_rand, key=len) if components_rand else set()
    G_random_lcc = G_random.subgraph(largest_cc_rand).copy()
    
    if nx.is_connected(G_random_lcc) and G_random_lcc.number_of_nodes() > 1:
        rand_diameter = nx.diameter(G_random_lcc)
        rand_avg_shortest = nx.average_shortest_path_length(G_random_lcc)
    else:
        rand_diameter = None
        rand_avg_shortest = None
    rand_clustering = nx.average_clustering(G_random)
    
    print("\nRandom Network Metrics:")
    print("Random graph average clustering coefficient:", rand_clustering)
    if rand_diameter is not None:
        print("Random graph diameter:", rand_diameter)
    if rand_avg_shortest is not None:
        print("Random graph average shortest path length:", rand_avg_shortest)
else:
    print("Random network generation failed.")
```

```python
suspicious_editors = ['Liuxinyu970226', 'M2k~dewiki']
abnormal_scores = {'Liuxinyu970226': 20, 'M2k~dewiki': 32}


risk_scores = {}
for editor in G.nodes():
    if editor in suspicious_editors:
        continue
    risk = 0
    for s in suspicious_editors:
        try:
            #  editor  s 
            d = nx.shortest_path_length(G, source=editor, target=s)
            risk += abnormal_scores[s] / d  # 
        except nx.NetworkXNoPath:
            continue
    risk_scores[editor] = risk

# 
priority_list = sorted(risk_scores.items(), key=lambda x: x[1], reverse=True)
print("Priority List (Editor, Risk Score):")
for editor, score in priority_list[:10]:
    print(f"{editor}: {score:.2f}")
```
